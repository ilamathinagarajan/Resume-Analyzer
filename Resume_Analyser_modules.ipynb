```python
#Extract Skilss From Resume and Job URL
#pip install requests bs4 nltk python-docx PyPDF2 pytesseract Pillow
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import download
import os
from docx import Document
from PyPDF2 import PdfReader
import pytesseract
from PIL import Image

# Set up NLTK data path
nltk.data.path.append(os.path.join(os.getcwd(), 'nltk_data'))

# Download required NLTK resources if not already present
try:
    stopwords.words('english')
    word_tokenize('test')
except LookupError:
    download('punkt')
    download('stopwords')

# Define the skill list
skills_list = ['Agile', 'Azure', 'DevOps', 'Machine Learning', 'Data Science', 'Deep Learning', 'Data Analysis',
               'Data Visualization', 'Cloud Computing', 'Big Data', 'Web Development', 'Software Development',
               'Mobile App Development', 'Frontend Development', 'Backend Development', 'Full Stack Development',
               'REST API', 'GraphQL', 'Database Management', 'Microservices', 'Docker', 'Kubernetes', 'CI/CD',
               'Scrum', 'Version Control (Git)', 'GitHub', 'GitLab', 'Automation', 'Test Automation', 'UI/UX Design',
               'Cybersecurity', 'Blockchain', 'Artificial Intelligence', 'Natural Language Processing (NLP)', 'TensorFlow',
               'Scikit-learn', 'AWS', 'Azure', 'Google Cloud', 'Azure DevOps', 'Postman', 'Jenkins', 'CI/CD Pipeline',
               'Cloud Security', 'Database Design', 'NoSQL', 'Relational Databases', 'MongoDB', 'MySQL', 'PostgreSQL',
               'Oracle', 'SQLAlchemy', 'Web Scraping', 'SEO', 'Selenium', 'Tableau', 'Power BI', 'Business Intelligence',
               'Game Development', 'API Development', 'Project Management', 'Agile Methodology', 'Kanban', 'Jira', 'Lean',
               'Software Testing', 'Unit Testing', 'JUnit', 'Mockito', 'JUnit Testing', 'Test-Driven Development (TDD)',
               'Microservices Architecture', 'CI/CD Pipelines', 'Cloud Infrastructure', 'Serverless Architecture', 'Containerization',
               'Infrastructure as Code', 'ElasticSearch', 'Nginx', 'Apache', 'JSP', 'JUnit', 'Maven', 'Redux', 'React', 'Vue.js',
               'Angular', 'Flutter', 'Xamarin', 'Laravel', 'Spring Framework', 'Spring Boot', 'FastAPI', 'Flask', 'Express.js',
               'Django', 'Node.js', 'Ruby on Rails', 'SaaS', 'Game Engines (Unity, Unreal Engine)', 'Docker Compose', 'Terraform',
               'Chef', 'Puppet', 'Vagrant', 'NGINX', 'Apache Kafka', 'Service-Oriented Architecture (SOA)', 'AWS Lambda',
               'WebSockets', 'Firebase', 'Vagrant', 'MongoDB', 'JPA', 'Hibernate', 'Figma', 'Photoshop', 'Illustrator',
               'UI/UX Design', 'Wireframing', 'Sketch', 'Zeplin', 'Canva', 'PowerShell', 'Terraform', 'Solidity', 'Automated Deployment',
               'Jenkins CI/CD', 'Ansible', 'Serverless Framework','Python', 'Java', 'C#', 'C++', 'JavaScript', 'TypeScript', 'Ruby', 'PHP',
               'Swift', 'Kotlin', 'Go', 'Rust', 'Perl', 'Objective-C', 'SQL', 'R', 'HTML', 'CSS', 'Sass', 'VHDL', 'MATLAB', 'Lua',
               'Shell Scripting', 'Dart', 'ActionScript', 'Scala', 'Haskell', 'Elixir', 'F#', 'Visual Basic', 'SQL Server', 'PL/SQL', 'T-SQL']

# Function to preprocess text
def preprocess_text(text):
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    return filtered_words

# Function to extract skills from text
def extract_skills_from_text(text, skill_list):
    text_lower = text.lower()
    extracted_skills = [skill for skill in skill_list if skill.lower() in text_lower]
    return extracted_skills

# Function to extract text from DOCX file
def extract_text_from_docx(docx_path):
    doc = Document(docx_path)
    return '\n'.join([paragraph.text for paragraph in doc.paragraphs])

# Function to extract text from PDF file
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    return '\n'.join(page.extract_text() for page in reader.pages)

# Function to extract text from TXT file
def extract_text_from_txt(txt_path):
    with open(txt_path, 'r', encoding='utf-8') as file:
        return file.read()

# Function to extract text from an image file
def extract_text_from_image(image_path):
    return pytesseract.image_to_string(Image.open(image_path))

# Main function to handle different file types
def extract_skills(file_path):
    file_extension = os.path.splitext(file_path)[1].lower()
    
    if file_extension == '.docx':
        text = extract_text_from_docx(file_path)
    elif file_extension == '.pdf':
        text = extract_text_from_pdf(file_path)
    elif file_extension == '.txt':
        text = extract_text_from_txt(file_path)
    elif file_extension in ['.jpg', '.jpeg', '.png']:
        text = extract_text_from_image(file_path)
    else:
        raise ValueError(f"Unsupported file type: {file_extension}")
    
    skills = extract_skills_from_text(text, skills_list)
    return skills

# Function to extract skills from a job description URL (Web scraping)
def extract_skills_from_job_description(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code != 200:
        print("Failed to retrieve the web page.")
        return []
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the job description text (you may need to customize this depending on the site structure)
    job_text = soup.get_text()
    
    # Preprocess the job text to lowercase for easy matching
    processed_job_text = preprocess_text(job_text)
    
    # Extract skills from the job description
    extracted_skills = extract_skills_from_text(job_text, skills_list)
    
    return extracted_skills
    
# Example usage
if __name__ == '__main__':
    file_path = 'C:\\Users\\Aswathy\\OneDrive\\Desktop\\Resume\\Aswathy_Resume.pdf'
    try:
        extracted_skills = extract_skills(file_path)
        print(f"Extracted Skills: {', '.join(extracted_skills)}")
    except Exception as e:
        print(f"Error: {e}")

     # Example usage for extracting skills from a job description URL
    job_url = "https://internshala.com/job/details/webflow-developer-job-in-pune-at-flytbase-labs1730810963?utm_source=cp_link&referral=web_share"  # Replace with your job description URL
    job_skills = extract_skills_from_job_description(job_url)
    
    # Print the extracted skills from the job description
    print(f"Skills found in job description: {', '.join(job_skills)}")
```

```python
import re
import nltk
import requests
import pdfplumber
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from tkinter import Tk, filedialog  # To open a file dialog for choosing resume
from bs4 import BeautifulSoup

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Step 1: Extract text from resume PDF
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text() or ''
    return text

# Step 2: Preprocess text (both resume and job description)
def preprocess_text(text):
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  # Lowercase
    tokens = word_tokenize(text)  # Tokenization
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]  # Stopword removal
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatization
    return ' '.join(tokens)

# Step 3: Extract skills from job description (dummy example, modify if needed)
def extract_skills_from_job_description(text):
    skills = ["python", "machine learning", "deep learning", "java", "sql",
              "cloud", "data analysis", "c++", "data visualization", "aws",
              "azure", "devops", "javascript", "html", "css", "react",
              "docker", "kubernetes", "linux", "tensorflow", "pandas",
              "scikit-learn", "excel", "power bi", "matlab", "r programming",
              "Bubble.io", "Express.js", "MongoDB"]  # Example skill list
    extracted_skills = [skill for skill in skills if skill in text.lower()]
    return extracted_skills

# Step 4: Extract skills from text
def extract_skills_from_text(text):
    skills = ["python", "machine learning", "deep learning", "java", "sql",
              "cloud", "data analysis", "c++", "data visualization", "aws",
              "azure", "devops", "javascript", "html", "css", "react",
              "docker", "kubernetes", "linux", "tensorflow", "pandas",
              "scikit-learn", "excel", "power bi", "matlab", "r programming",
              "Bubble.io", "Express.js", "MongoDB"]  # Example skill list
    extracted_skills = [skill for skill in skills if skill in text.lower()]
    return extracted_skills

# Step 5: Compare skills from resume and job description
def compare_skills(resume_skills, job_skills):
    matched_skills = set(resume_skills).intersection(set(job_skills))
    missing_skills = set(job_skills).difference(set(resume_skills))
    return matched_skills, missing_skills

# Main execution
def main():
    # Step 1: Open file dialog to select resume PDF
    root = Tk()
    root.withdraw()  # Don't need the main Tkinter window
    print("Please select the resume PDF file.")
    resume_path = filedialog.askopenfilename(filetypes=[("PDF files", "*.pdf")])

    if not resume_path:
        print("No file selected, exiting.")
        return

    # Extract text from the selected PDF file
    resume_text = extract_text_from_pdf(resume_path)
    cleaned_resume_text = preprocess_text(resume_text)

    # Step 2: Get job description URL input from the user
    job_url = input("Enter the job description URL: ").strip()

    # Step 3: Web scrape the job description
    response = requests.get(job_url)
    job_description_text = response.text
    cleaned_job_description_text = preprocess_text(job_description_text)

    # Step 4: Extract skills from resume and job description
    resume_skills = extract_skills_from_text(cleaned_resume_text)
    job_skills = extract_skills_from_text(cleaned_job_description_text)

    # Display extracted skills
    print(f"Skills extracted from resume: {resume_skills}")
    print(f"Skills required for the job description: {job_skills}")

    # Step 5: Compare skills and display the output
    matched_skills, missing_skills = compare_skills(resume_skills, job_skills)

    # Display matched skills
    print(f"Matched Skills: {matched_skills}")
    print(f"Missing Skills: {missing_skills}")

if __name__ == '__main__':
    main()
#C:\Users\Aswathy\OneDrive\Desktop>python missing_skill.py
#Please select the resume PDF file.
#Enter the job description URL: https://internshala.com/job/details/web-developer-job-in-delhi-at-one2all-recharge-private-limited1731580291
#Skills extracted from resume: ['python', 'java', 'sql', 'cloud', 'data analysis', 'azure', 'javascript', 'html', 'excel']
#Skills required for the job description: ['python', 'machine learning', 'java', 'sql', 'aws', 'javascript', 'html', 'css', 'react', 'excel']
#Matched Skills: {'python', 'excel', 'java', 'html', 'sql', 'javascript'}
#Missing Skills: {'aws', 'css', 'machine learning', 'react'}
```
